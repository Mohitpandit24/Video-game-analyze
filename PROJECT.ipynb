```python
def count_plot(x, data, ax):
    sns.countplot(x = x, data = data, ax = ax)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    
fig, axes=plt.subplots(3, figsize = (15,15))
count_plot("Genre" , df, axes[0])
count_plot("Year" , df, axes[1])
count_plot("Platform" , df, axes[2])

axes[0].set_title("Genre name")
axes[1].set_title("Year count")
axes[2].set_title("platform count")

plt.tight_layout()
plt.show()
```

# What are the most popular genres over time?

```python
dfm=df.groupby(["Genre"], as_index=False)["Global_Sales"].sum().sort_values(by="Global_Sales", ascending=False).head(10)
sns.set(rc={"figure.figsize":(10,5)})
sns.barplot(data=dfm, x ="Genre", y="Global_Sales")
plt.xlabel("Genre")
plt.ylabel("Global_Sales")
plt.title("Top 10 Genres by Global Sales")
plt.show()
```

# Are certain genres more popular in specific regions?

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: NA_Sales
data_na = df.groupby(["Genre"], as_index=False)["NA_Sales"].sum().sort_values(by="NA_Sales", ascending=False).head(10)
sns.barplot(data=data_na, x="Genre", y="NA_Sales", ax=axes[0, 0])
axes[0, 0].set_title("Genre Sales in North America")
axes[0, 0].tick_params(axis='x', rotation=90)

# Plot 2: EU_Sales
data_eu = df.groupby(["Genre"], as_index=False)["EU_Sales"].sum().sort_values(by="EU_Sales", ascending=False).head(10)
sns.barplot(data=data_eu, x="Genre", y="EU_Sales", ax=axes[0, 1])
axes[0, 1].set_title("Genre Sales in Europe")
axes[0, 1].tick_params(axis='x', rotation=90)

# Plot 3: JP_Sales
data_jp = df.groupby(["Genre"], as_index=False)["JP_Sales"].sum().sort_values(by="JP_Sales", ascending=False).head(10)
sns.barplot(data=data_jp, x="Genre", y="JP_Sales", ax=axes[1, 0])
axes[1, 0].set_title("Genre Sales in Japan")
axes[1, 0].tick_params(axis='x', rotation=90)

# Plot 4: Other_Sales
data_other = df.groupby(["Genre"], as_index=False)["Other_Sales"].sum().sort_values(by="Other_Sales", ascending=False).head(10)
sns.barplot(data=data_other, x="Genre", y="Other_Sales", ax=axes[1, 1])
axes[1, 1].set_title("Genre Sales in Other Regions")
axes[1, 1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

# What are the best-selling video games globally?

```python
dfn=df.groupby(["Game"], as_index=False)["Global_Sales"].sum().sort_values(by="Global_Sales", ascending=False).head(5)
sns.set(rc={"figure.figsize":(10,5)})
sns.histplot(data=dfn, x="Game", y="Global_Sales")
plt.show()
```

# Which region contributes the most to global sales (NA, EU, JP, or other)?

```python
region_sales={"North America":df["NA_Sales"].sum(),
             "Europe":df["EU_Sales"].sum(),
             "Japan":df["JP_Sales"].sum(),
             "Other": df['Other_Sales'].sum()}
max_region=max(region_sales, key=region_sales.get)
print("Total Sales by Region:", region_sales)
print(f"The region contributing the most to global sales is: {max_region}")
```

# How have global sales of video games changed over the years?

```python
global_sales_year = df.groupby(df['Year'])['Global_Sales'].sum()
fig = plt.figure(figsize=(15,7))
plt.plot(global_sales_year, '-')
plt.title("Global Sales Over Year")
plt.xlabel("Year")
plt.ylabel("Global Sales")
plt.show()
```

# Which publishers dominate sales in different regions?

```python
region_sales = df.groupby("Publisher")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].sum()

top_publishers = region_sales.idxmax()

print("Top Publishers by Region:")
print(top_publishers)

fig, ax = plt.subplots(2, 2, figsize=(12, 10))

regions = ["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]
titles = ["North America", "Europe", "Japan", "Other Regions"]
colors = ["skyblue", "salmon", "lightgreen", "gold"]

for i, region in enumerate(regions):
    region_data = region_sales[region].sort_values(ascending=False).head(10)  # Top 10 publishers
    ax[i // 2, i % 2].bar(region_data.index, region_data.values, color=colors[i])
    ax[i // 2, i % 2].set_title(f"Top Publishers in {titles[i]}")
    ax[i // 2, i % 2].set_xticklabels(region_data.index, rotation=45, ha="right")
    ax[i // 2, i % 2].set_ylabel("Sales (in millions)")
    
plt.tight_layout()
plt.show()
```
